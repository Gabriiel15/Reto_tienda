### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
import matplotlib.pyplot as plt
```

#1. Análisis de facturación



```python
venta_tienda = tienda['Precio'].sum()
print(f'La venta bruta de la primer tienda es de: {venta_tienda}')
costo_envio = tienda['Costo de envío'].sum()
print(f'El costo de envio total de la primer tienda es: {costo_envio}')
venta_total_tienda = venta_tienda - costo_envio
print(f'La venta total de la primer tienda es: {venta_total_tienda}')

print()

venta_tienda = tienda2['Precio'].sum()
print(f'La venta bruta de la segunda tienda es de: {venta_tienda}')
costo_envio = tienda2['Costo de envío'].sum()
print(f'El costo de envio total de la segunda tienda es: {costo_envio}')
venta_total_tienda2 = venta_tienda - costo_envio
print(f'La venta total de la segunda tienda es: {venta_total_tienda2}')

print()

venta_tienda = tienda3['Precio'].sum()
print(f'La venta bruta de la tercer tienda es de: {venta_tienda}')
costo_envio = tienda3['Costo de envío'].sum()
print(f'El costo de envio total de la tercer tienda es: {costo_envio}')
venta_total_tienda3 = venta_tienda - costo_envio
print(f'La venta total de la tercer tienda es: {venta_total_tienda3}')

print()

venta_tienda = tienda4['Precio'].sum()
print(f'La venta bruta de la cuarta tienda es de: {venta_tienda}')
costo_envio = tienda4['Costo de envío'].sum()
print(f'El costo de envio total de la cuarta tienda es: {costo_envio}')
venta_total_tienda4 = venta_tienda - costo_envio
print(f'La venta total de la cuarta tienda es: {venta_total_tienda4}')
```

```python

# Supongamos que tienes tus DataFrames: tienda, tienda2, tienda3, tienda4
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

def calcular_ventas_por_categoria(tienda):
    venta_categoria = tienda.groupby('Categoría del Producto')['Precio'].sum()
    return venta_categoria.sort_values(ascending=False)

# Crear un diccionario para almacenar las ventas por categoría de cada tienda
ventas_por_categoria = {}

for nombre, tienda in zip(nombres_tiendas, tiendas):
    ventas_por_categoria[nombre] = calcular_ventas_por_categoria(tienda)

# Imprimir los resultados
for nombre, ventas in ventas_por_categoria.items():
    print(f"Ventas por categoría en {nombre}:")
    print(ventas)
    print()
```

```python
def graficar_ventas_pastel(venta_categoria, nombre_tienda):
    plt.pie(venta_categoria, labels=venta_categoria.index, autopct='%1.1f%%')
    plt.title(f'Ventas por categoría {nombre_tienda}')
    plt.axis('equal')
    plt.show()

graficar_ventas_pastel(venta_categoria, 'Tienda 1')
graficar_ventas_pastel(venta_categoria2, 'Tienda 2')
graficar_ventas_pastel(venta_categoria3, 'Tienda 3')
graficar_ventas_pastel(venta_categoria4, 'Tienda 4')
```
